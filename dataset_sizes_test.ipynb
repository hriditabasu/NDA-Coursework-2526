<h3>Determine File Sizes (For Selection)

```python
import os

files = [
    "ADMINISTRATORS.csv",
    "BOT_REQUESTS.csv",
    "INTERWIKI_CONFLICT.csv",
    "ITEMS.csv",
    "OTHERS.csv",
    "PROJECT_CHAT.csv",
    "PROPERTIES.csv",
    "PROPERTIES_FOR_DELETION.csv",
    "PROPERTY_PROPOSAL.csv",
    "REPORT_A_TECHNICAL_PROBLEM.csv",
    "REQUEST_A_QUERY.csv",
    "REQUEST_FOR_DELETION.csv",
    "REQUESTS_FOR_COMMENT.csv",
    "REQUESTS_FOR_PERMISSIONS.csv",
    "USERS.csv",
    "WIKIPROJECTS.csv"

    # add all 16 here
]
filesizes = []
for file in files:
    size = os.path.getsize("datasets/" + file)
    filesizes.append((file, size))

files_sorted = sorted(filesizes, key=lambda x: x[1])
for f in files_sorted:
    print(f)
```

The 3 datasets (selected using file size as a proxy for network scale in accordance with
coursework instructions) are:
- Small: BOT_REQUESTS.csv (263 KB)
- Medium: ADMINISTRATORS.csv (1.75 MB)
- Large: USERS.csv (10 MB)
